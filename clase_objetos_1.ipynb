```python
class Gato:
  numero_patas = 4
  def __init__(self,raza,nombre,edad):
    self.raza = raza
    self.nombre = nombre
    self.edad = edad
  def maullar(self):
    print("El gato esta atento: Miau !!!")
  def get_raza(self):
    print("La raza del gato es =", self.raza)
  def get_nombre(self):
    print("El nombre del gato es =", self.nombre)
  def get_edad(self):
    print("La edad del gato es =", self.edad)

g = Gato("Persa", "Urano", 2)
g.maullar()
g.get_raza()
g.get_nombre()
g.get_edad()

print("Número de patas del gato es =", g.numero_patas)
```

```python
class Moto:
  numero_ruedas = 2
  def __init__(self,marca,tipo,año):
    self.marca = marca
    self.tipo = tipo
    self.año = año
  def get_marca(self):
    print("La marca de la moto es =", self.marca)
  def get_tipo(self):
    print("El tipo es =", self.tipo)
  def get_año(self):
    print("El año es =", self.año)

m = Moto("Ducati", "Panigale V4", 2022)
m.get_marca()
m.get_tipo()
m.get_año()

print("Número de ruedas =", m.numero_ruedas)
```
